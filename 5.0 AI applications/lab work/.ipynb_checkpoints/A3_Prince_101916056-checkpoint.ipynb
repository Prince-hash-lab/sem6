```python
import pandas as pd

data = pd.read_pickle('dtm.pkl')
data.head()
```

```python
data = data.transpose() 
data.head()
```

```python
#Find and print the top 30 words said by each comedian
top_30 = {}

for c in data.columns:
    
    top = data[c].sort_values(ascending=False).head(30)
    
    top_30[c]= list(zip(top.index, top.values))

#top_30
```

```python
#this will print top 30 words used by every comedian with their names
for comedian, top_words in top_30.items():
  print(comedian + ':')
  print(', '.join([word for word, count in top_words[0:29]]))
  print(' ')
  print(' ')
```

```python
# this will give most common top words used by all comedians 
from collections import Counter

common_words = []
for comedian in data.columns:
    top = [word for (word, count) in top_30[comedian]]
    for t in top:
        common_words.append(t)   
Counter(common_words).most_common()
common_words
```

```python
# counter package will give freq of most common words using this
Counter(common_words).most_common()
```

```python
# aggregate the list of the common words and more than half of the comedians have it as their top word
new_stop_words = [word for word, count in Counter(common_words).most_common() if count > 6]
new_stop_words
```

```python
# Let's update our document-term matrix with the new list of stop words
# we have used pickle here 

import pickle
from sklearn.feature_extraction import text 
from sklearn.feature_extraction.text import CountVectorizer

# Read in cleaned data

data_clean = pd.read_pickle('data_clean.pkl')

# Add new stop words

stop_words = text.ENGLISH_STOP_WORDS.union(new_stop_words)

# Recreate document-term matrix

cv = CountVectorizer(stop_words=stop_words)
data_cv = cv.fit_transform(data_clean.transcript)
data_stop = pd.DataFrame(data_cv.toarray(), columns=cv.get_feature_names())
data_stop.index = data_clean.index

# Pickle it for later use
pickle.dump(cv, open("cv_stop.pkl", "wb"))
data_stop.to_pickle("dtm_stop.pkl")
```

```python
#  word clouds is created here
from wordcloud import WordCloud

wc = WordCloud(stopwords=stop_words, background_color="white")
```

```python
import matplotlib.pyplot as plt

names = ['Ali', 'Anthony', 'Bill', 'Bo', 'Dave', 'Hasan','Jim', 'Joe', 'John', 'Louis', 'Mike', 'Ricky']

# subplots for each comedian is created here 
for index, comedian in enumerate(data.columns):
    wc.generate(data_clean.transcript[comedian])
    
    plt.subplot(3, 4, index+1)
    plt.imshow(wc, interpolation="bilinear")
    plt.axis("off")
    plt.title(names[index])
    
plt.show()
```

```python
# Find the number of unique words that each comedian uses

# Identify the non-zero items in the document-term matrix, meaning that the word occurs at least once
unique_list = []
for comedian in data.columns:
    uniques = data[comedian].to_numpy().nonzero()[0].size
    unique_list.append(uniques)

# Create a new dataframe that contains this unique word count
data_words = pd.DataFrame(list(zip(names, unique_list)), columns=['comedian', 'unique_words'])

data_unique_sort = data_words.sort_values(by='unique_words')

data_unique_sort
```

```python
# Calculate the words per minute of each comedian

# Find the total number of words that a comedian uses
total_words = []
for comedian in data.columns:
    totals = sum(data[comedian])
    total_words.append(totals)

# add some columns to our dataframe: 'total_words', 'run_times' and 'words_per_minute'
data_words['total_words'] = total_words
data_words['run_times'] = [60, 60, 60, 60, 60, 60, 60, 60, 60, 60, 60, 60]
data_words['words_per_minute'] = data_words['total_words'] / data_words['run_times']

# Sort the dataframe by words per minute to see who talks the slowest and fastest
data_sort = data_words.sort_values(by='words_per_minute')
data_sort
```

```python
import numpy as np
# first
y_pos = np.arange(len(data_words))

plt.subplot(1, 2, 1)

plt.barh(y_pos, data_unique_sort.unique_words, align='center')
plt.yticks(y_pos, data_unique_sort.comedian)
plt.title('Number of Unique Words', fontsize=20)

#second
plt.subplot(1, 2, 2)
plt.barh(y_pos, data_sort.words_per_minute, align='center')
plt.yticks(y_pos, data_sort.comedian)
plt.title('Number of Words Per Minute', fontsize=20)

plt.tight_layout()
plt.show()
```
